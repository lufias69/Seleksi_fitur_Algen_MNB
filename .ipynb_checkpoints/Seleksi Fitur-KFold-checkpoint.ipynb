```python
import pandas as pd
import numpy as np
from Lib import Algen_lib as lib
from Lib import AlgenFS_KF as fs
```

```python
#Data Latih
df = pd.read_excel('data/sms_clean_norm_+_new_data_3.xlsx')

corpusA = df.sms.tolist()
labelA = df.label.tolist()
len(df)
```

```python
fitur = lib.feature_extraction(corpusA)
```

```python
jumlah_populasi = 1000
jumlah_titik_co = 10 #jumlah titik point crossover
prob_mutasi = 0.2 #probabilitas permutasi
K=25
metode = "tf"
alpha = 1
target = 0.99 #target akurasi
panjang_fitur = len(fitur)
```

```python
hasil_seleksi = fs.features_selection(corpusA,labelA, fitur=fitur, jumlah_populasi=jumlah_populasi,
                                   jumlah_titik=jumlah_titik_co, prob_mutasi=prob_mutasi,
                                   K=K, alpha=alpha, target=target, 
                                   panjang_fitur=panjang_fitur,
                                   metode=metode, additional_child=True)
```

```python
print(hasil_seleksi['best_score'])
print(hasil_seleksi['best_fitur'])
print(hasil_seleksi['good_fitur'])
```

```python
lib.fitness_kf(corpusA, labelA, hasil_seleksi['best_fitur_bin'], fitur, alpha = alpha, metode = metode, K=K)
```

```python
lib.fitness_kf(corpusA, labelA, hasil_seleksi['good_fitur_bin'], fitur, alpha = alpha, metode = metode, K=K)
```
